```python
import subprocess
import ollama
import re
from pydub import AudioSegment
from pydub.playback import play
from IPython.display import Audio

def run_f5_tts_inference(model="F5-TTS", ref_audio="ref_audio.wav",
                         ref_text=".",
                         gen_text="Some text you want the TTS model to generate for you.",
                         config_file=None):
    cmd = [
        "f5-tts_infer-cli",
        "--model", model,
        "--ref_audio", ref_audio,
        "--ref_text", ref_text,
        "--gen_text", gen_text
    ]

    if config_file:
        cmd.extend(["-c", config_file])
    print("Running command:", " ".join(cmd))

    try:
        result = subprocess.run(cmd, check=True, capture_output=True, text=True)
        print("Inference completed successfully.")
        print("Output:\n", result.stdout)
    except subprocess.CalledProcessError as err:
        print("Error during inference:")
        print(err.stderr)


```

```python
def pirateKing(reference_audio_path, reference_text, custom_config, generation_text, model):
    response = ollama.chat(model=model, messages=[
        {
        'role' : 'user',
        'content' : generation_text+" .And also write only as if you are talking to me as a British person from 1900's and do not add anything extra"
        },
    ])
    res = response['message']['content']
    res = re.sub(r'<think>.*?</think>', '', res, flags=re.DOTALL)
    run_f5_tts_inference(ref_audio=reference_audio_path, ref_text=reference_text,
                         gen_text=res, config_file=custom_config)
```

```python
import os
import ipywidgets as widgets
from IPython.display import display, Audio, clear_output

header = widgets.HTML(
    value="""
    <h2 style='text-align:center; color:#2C3E50; font-family: Arial, sans-serif;'>
        üéµ Audio Generation Interface üéµ
    </h2>
    <p style='text-align:center; color:#7F8C8D; font-family: Arial, sans-serif;'>
        Generate high-quality audio from text and reference inputs.
    </p>
    """
)


model_name = widgets.Text(
    placeholder=model,
    description="üõ† Model:",
    layout=widgets.Layout(width="90%", padding="8px", border="2px solid #BDC3C7", border_radius="5px"),
    style={"description_width": "initial"}
)

reference_audio_path = widgets.Text(
    placeholder="Enter reference audio path...",
    description="üé§ Audio Path:",
    layout=widgets.Layout(width="90%", padding="8px", border="2px solid #BDC3C7", border_radius="5px"),
    style={"description_width": "initial"}
)

generation_text = widgets.Textarea(
    placeholder="Enter generation text here...",
    description="üìù Text:",
    layout=widgets.Layout(
        width="90%",
        height="100px",
        padding="10px",
        #border="10px solid #BDC3C7",
        #border_radius="15px",
    #    box_sizing="border-box"  
    ),
    style={"description_width": "initial"}
)



config_section = widgets.VBox(
    layout=widgets.Layout(margin="15px 0px 10px 0px")
)

custom_config_options = ["None", "Custom"]
custom_config_dropdown = widgets.Dropdown(
    options=custom_config_options,
    value="None",
    description="‚öôÔ∏è Config:",
    layout=widgets.Layout(width="50%"),
    style={"description_width": "initial"}
)

custom_config_text = widgets.Text(
    placeholder="Enter configuration file path...",
    description="üìÇ File Path:",
    layout=widgets.Layout(width="90%", padding="8px", border="2px solid #BDC3C7", border_radius="5px"),
    style={"description_width": "initial"},
    disabled=True
)
custom_config_text.layout.display = "none"

def on_custom_config_change(change):
    if change["new"] == "Custom":
        custom_config_text.disabled = False
        custom_config_text.layout.display = "block"
    else:
        custom_config_text.disabled = True
        custom_config_text.value = ""
        custom_config_text.layout.display = "none"

custom_config_dropdown.observe(on_custom_config_change, names="value")
config_section.children = [custom_config_dropdown, custom_config_text]


save_audio_path = widgets.Text(
    placeholder="Enter path to save generated audio file...",
    description="üíæ Save Path:",
    layout=widgets.Layout(width="90%", padding="8px", border="2px solid #BDC3C7", border_radius="5px"),
    style={"description_width": "initial"}
)


inference_button = widgets.Button(
    description="üîÑ Run Inference",
    button_style="success",
    tooltip="Run inference to generate audio",
    layout=widgets.Layout(width="45%", height="100%", padding="10px", font_size="14px", border_radius="8px")
)
inference_button.icon = "üîÑ"  

play_button = widgets.Button(
    description="üîä Play Audio",
    button_style="info",
    tooltip="Play the generated audio file",
    layout=widgets.Layout(width="45%", height="100%", padding="10px", font_size="14px", border_radius="8px"),
    disabled=True
)
play_button.icon = "üîä"



message_output = widgets.Output()


def run(b):
    with message_output:
        message_output.clear_output()
        ref_audio = reference_audio_path.value.strip()
        text_input = generation_text.value.strip()
        config_choice = custom_config_dropdown.value
        config_path = custom_config_text.value.strip() if config_choice == "Custom" else None
        save_path = save_audio_path.value.strip()

        print(ref_audio)
        print(text_input)

        if not save_path:
            print("‚ö†Ô∏è Please provide a valid path to save the generated audio file.")
            return

        if os.path.exists(save_path):
            try:
                os.remove(save_path)
                print(f"üóë Deleted existing audio file at: {save_path}")
            except Exception as e:
                print(f"‚ùå Error deleting file at {save_path}: {e}")
                return

        try:

            pirateKing(ref_audio, "", None, text_input, model)
            print(f"‚úÖ Inference complete. Audio file generated at: {save_path}")
        except Exception as e:
            print(f"‚ùå Error generating audio file: {e}")
            return

        play_button.disabled = False


def play_audio(b):
    with message_output:
        message_output.clear_output()
        save_path = save_audio_path.value.strip()

        if not os.path.exists(save_path):
            print("‚ùå Audio file not found!")
            return

        try:
            audio = AudioSegment.from_file(save_path)
            play(audio)
            display(Audio(save_path, autoplay=True))

        except Exception as e:
            print(f"‚ùå Error playing audio: {e}")

inference_button.on_click(run)
play_button.on_click(play_audio)


button_box = widgets.HBox(
    [inference_button, play_button],
    layout=widgets.Layout(justify_content="center", gap="100px", padding="10px")
)

ui_container = widgets.VBox(
    [
        header,
        model_name,
        reference_audio_path,
        generation_text,
        config_section,
        save_audio_path,
        button_box,
        message_output,
    ],
    layout=widgets.Layout(
        align_items="center",
        padding="20px",
        border="2px solid #BDC3C7",
        border_radius="10px",
        box_shadow="4px 4px 15px rgba(0, 0, 0, 0.1)",
        width="100%",
        background_color="#F8F9FA"
    )
)

display(ui_container)
#/content/tests/infer_cli_basic.wav
#/tests/infer_cli_basic.wav
```
